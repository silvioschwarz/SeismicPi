```python
data = pd.read_csv('functions/logs/log-2019-05-23-06:45:43.csv')
data.columns = ['time',
                'accx','accy','accz', 
                'magx', 'magy', 'magz',
                'gyrox','gyroy', 'gyroz']

data.head()
```

```python
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from datetime import datetime

fig, ax = plt.subplots(figsize=(14,10))
myFmt = DateFormatter("%M:%S.%f") 

x = [datetime.strptime(d,"%Y-%m-%d-%H:%M:%S.%f") for d in data.time]


ax.xaxis_date()
ax.xaxis.set_major_formatter(myFmt); 

ax.plot(x,data.accx)

ax.set_xlim(datetime(2019, 5, 23, 6, 48, 43),
            datetime(2019, 5, 23, 6, 49, 4, 266422))
plt.grid(True)
plt.show()
```
