```python
data = pd.read_csv('logs/log-2019-05-23-09:13:09.csv')
data.columns = ['time',
                'accx','accy','accz', 
                'magx', 'magy', 'magz',
                'gyrox','gyroy', 'gyroz']

data.head()
```

```python
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from datetime import datetime

fig, (ax1,ax2,ax3) = plt.subplots(3,1,figsize=(24,8))
myFmt = DateFormatter("%M:%S.%f") 

x = [datetime.strptime(d,"%Y-%m-%d-%H:%M:%S.%f") for d in data.time]


#ax1.xaxis_date()
#ax.xaxis.set_major_formatter(myFmt); 


ax1.plot(x,data.accx)
plt.grid(True)
ax2.plot(x,data.accy)
plt.grid(True)
ax3.plot(x,data.accz)
plt.grid(True)

plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9,hspace = 0.6)
#ax.xaxis.tick_top()
#ax.xaxis.set_label_position("top")
#ax.set_xlim(-5,10)
#ax.set_ylim(-15,2)
#ax1.set_aspect('equal')
ax1.set_xlabel("X [km]")
ax1.set_ylabel("Acceleration [g]")
ax1.set_title("X")

ax2.set_xlabel("X [km]")
ax2.set_ylabel("Acceleration [g]")
ax2.set_title("Y")
plt.grid(True)
ax3.set_xlabel("time")
ax3.set_ylabel("Acceleration [g]")
ax3.set_title("Z")

#ax.set_xlim(datetime(2019, 5, 23, 6, 48, 43),
#            datetime(2019, 5, 23, 6, 49, 4, 266422))

plt.show()
```
